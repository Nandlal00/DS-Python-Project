# 2. Create a new text file for each of the characters with their name and store the unique words 

```python
#create dictionary to store the data
data = {}
```

```python
for speaker in character:
    data[speaker] = set()
```

```python
data
```

```python
for dailogues in text:
    if dailogues != '\n':
        dailogues = dailogues.split(":")
        
        speaker = dailogues[0]
        
        dailoguesofspeaker = dailogues[1]
        
        data[speaker].update(dailoguesofspeaker.split())
```

```python
for speaker in data.keys():
    #create a text file for every speaker in write mode
    text_file = open(f'{speaker}.txt', mode = "w+")
    #store all dailogues in list
    words = list(data[speaker])
    
    words = [word.strip('.,!;()[]?') for word in words]
    words = [word.lower() for word in words]
    
    text_file.writelines("\n".join(words))
```
